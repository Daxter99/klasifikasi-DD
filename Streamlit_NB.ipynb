```python
import pandas as pd
```

```python
dd_prediction = pd.read_csv('data.csv')
```

```python
data = dd_prediction.copy()
```

```python
target = 'Label'
```

```python
encode = ['Jenis Kelamin', 'Ruam Kulit', 'Manifestasi perdarahan', 'Kegagalan Sirkulasi', 'Syok berat', 'Uji Tokniket', 'Kebocoran Plasma', 'Pendarahan spontan']
```

```python
data
```

```python
for col in encode:
  dummy = pd.get_dummies(data[col], prefix=col)
  data = pd.concat([data, dummy], axis = 1)
  del data[col]
```

```python
data
```

```python
X = data.drop('Label', axis = 1)
Y = data['Label']
```

```python
Y
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X, Y)
```

```python
import pickle
pickle.dump(model, open('model_dd.pkl', 'wb'))
```
